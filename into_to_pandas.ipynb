## What's Pandas?

The pandas library is a popular open-source data manipulation and analysis tool for Python.

We use pandas in data science because it provides powerful tools for data manipulation, cleaning, analysis, and visualization, allowing us to work with tabular and structured data more efficiently and effectively.





````python
import pandas as pd
````

## What's a DataFrame?

A dataframe is a 2-dimensional labeled data structure in pandas that consists of rows and columns, similar to a spreadsheet or SQL table.

````python
data = {'Name': ['Bamboo', 'Ming Ming', 'Giant', 'Xiong Mao'],
        'Age': [4, 6, 3, 5],
        'Color': ['Black and White', 'Brown', 'White', 'Black'],
        'Favorite Food': ['Bamboo', 'Bamboo shoots', 'Apples', 'Fish'],
        'Weight (lbs)': [220, 180, 300, 250]}
````

````python
df = pd.DataFrame(data)
df
````

### What's a Series?

A series is a one-dimensional labeled data structure in pandas that can hold any data type, such as integers, floats, strings, and more

````python
df["Name"] # df.Name
````

### What's an Index?

A dataframe index is a unique identifier for each row in a pandas DataFrame, allowing for easy and efficient access and manipulation of data.

````python
df.index.values
````

## Data Cleaning

### Handling missing values

````python
# checking if it contains aby NULL values
df.isnull().any()
````

````python
# get the number of missing data points per column
missing_values_count = df.isnull().sum()
missing_values_count
````

````python
# remove all the rows that contain a missing value
df.dropna()
````

````python
# remove all columns with at least one missing value
columns_with_na_dropped = df.dropna(axis=1)

# just how much data did we lose?
print(f"Columns in original dataset: {df.shape[1]}")
print(f"Columns with na's dropped: {columns_with_na_dropped.shape[1]}")
````

````python
df.fillna(0)
````

### Replace values

````python
df['Targeted Sect if any'].value_counts()
````

````python
df['Targeted Sect if any'].replace(('shiite', 'Shiite/sunni'), ('Shiite', 'Shiite'), inplace  = True)
````

````python
df['Holiday Type'].unique()
````

````python

df['Holiday Type'] = df['Holiday Type'].str.lower()
df['Holiday Type'] = df['Holiday Type'].str.strip()
df['Holiday Type'].unique()
````

### Changing the Index

````python
df.index.values
df['S#'].is_unique

df.set_index('S#', inplace=True)

df.reset_index(drop=True, inplace=True)
````

## Data Aggregation

````python
df_holiday_types = df.groupby(['Holiday Type']).sum()
df_holiday_types.sort_values(["Killed Max"], ascending=False)
````

````python
df.groupby('Province')[['Killed Min', 'Injured Min']].sum()
````

````python
df.groupby('Province')[['Killed Min', 'Injured Min']].agg(['sum', 'mean'])
````

## Data Visualisation

````python
import matplotlib.pyplot as plt
````

````python
df['Target Type'].value_counts().plot.bar(figsize = (15, 5), color = 'pink')
plt.title('Targets set by the Terrorists', fontsize = 20)
plt.xlabel('Targets')
plt.ylabel('Count')
plt.show()
````

````python
df['Open/Closed Space'].replace(('open', 'closed', 'Open/Closed', 'Open '),('Open', 'Closed', 'Open', 'Open'), inplace = True)

df['Open/Closed Space'].value_counts().plot.pie(colors = ['red', 'grey'], explode = [0, 0.1])
plt.title('A pie Chart representing choice of terrorists for bombing', fontsize = 20)
plt.axis('off')
plt.legend()
plt.show()
````

### lambda

In Python, a lambda function is a small, anonymous function that can be defined without a name. Instead of using the def keyword to define a function with a name, you use the lambda keyword to define a function "on the fly".

lambda functions are often used as "throwaway" functions, when you need a quick function to perform a simple operation. They can also be used as arguments to other functions, such as the map(), filter(), and reduce() functions, which take functions as inputs.


````python
add = lambda x, y: x + y

result = add(2, 3)
print(result)  
````

### .apply()

In pandas, the apply() function is used to apply a function along a specific axis of a DataFrame. The apply() function takes a single argument, which is the function to be applied, and applies that function to each row or column of the DataFrame, depending on the value of the axis parameter.

The basic syntax for using the apply() function is:

```df.apply(func, axis=0)```


````python
df = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})

def sum_row(row):
    return row["A"] + row["B"]

df["sum"] = df.apply(sum_row, axis=1)
````
